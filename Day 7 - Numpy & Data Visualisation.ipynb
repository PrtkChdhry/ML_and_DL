## Data Visualisation ---

### Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
days = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
steps = [6745,4823,7215,1023,4503,2018,9703]
```

```python
plt.plot(steps)
```

```python
plt.plot(days,steps)
```

```python
plt.plot(days,steps,'o')
```

```python
plt.plot(days,steps,'o-')
```

```python
days = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
lws = [7245,4021,5361,8640,1203,9410,1569]  # last week steps
steps = [6745,4823,7215,1023,4503,2018,9703]  # this week steps
```

```python
plt.plot(days,steps,'o-g')
```

```python
plt.plot(days,lws,'o-m')
```

```python
plt.plot(days,steps,'o-g')
plt.plot(days,lws,'o-m')
plt.title('Steps Count - last week & this week')
plt.xlabel('Days of the week')
plt.ylabel('Steps walked')
```

```python
plt.plot(days,steps,'o-g')
plt.plot(days,lws,'o-m')
plt.title('Steps Count - last week & this week')
plt.xlabel('Days of the week')
plt.ylabel('Steps walked')
plt.grid(True)
```

```python
days = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
lws = [7245,4021,5361,8640,1203,9410,1569]  # last week steps
steps = [6745,4823,7215,1023,4503,2018,9703]  # this week steps
plt.plot(days,steps,'o-g')
plt.plot(days,lws,'o-m')
plt.title('Steps Count - last week & this week')
plt.xlabel('Days of the week')
plt.ylabel('Steps walked')
plt.grid(True)
plt.legend(['Current Week','Last Week'])
```

```python
plt.figure(figsize=(10,5))
age = [34,36,28,24,30]
name = ['VK','RS','SKY','IK','HP']
plt.bar(name,age)
```

```python
name = ['Prateek','Vani','Prachi','Rishika']
ot = [29,48,32,45]  
ct = [22,34,15,29]  
plt.bar(name,ot,label='Open Ticket',width=0.5,color='k')
plt.bar(name,ct,label='Closed Ticket',width=0.5,color='g')
plt.xlabel('Name')
plt.ylabel('Tickets')
plt.legend()
```

```python
age = [34,36,22,24,25]
name = ['VK','RS','SKY','IK','HP']
colors = ['g','r','b','k','m']
plt.pie(age,labels=name,colors=colors)
```
