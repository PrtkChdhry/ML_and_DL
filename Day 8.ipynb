### Pie chart

```python
data = [2,6,4,3,10]
plt.pie(data)
```

```python
data = [2,6,4,3,10]
plt.pie(data)
plt.show()  # it will not show all those above texts
```

```python
plt.pie(data,[0,0.1,0,0.3,0])
```

```python
plt.pie(data,[0,0.1,0,0.3,0],labels=['t1','t2','t3','t4','t5'])
```

```python
plt.pie(data,[0,0.1,0,0.3,0],labels=['t1','t2','t3','t4','t5'],autopct='%1.1f%%')
```

```python
plt.pie(data,[0,0.1,0,0.3,0],labels=['t1','t2','t3','t4','t5'],autopct='%1.1f%%',shadow=True)
```

```python
plt.pie(data,[0,0.1,0,0.3,0],labels=['t1','t2','t3','t4','t5'],autopct='%1.1f%%',shadow=True,colors=['yellow','red','pink','cyan','purple'])
```

```python
plt.pie(data,[0,0.1,0,0.3,0],labels=['t1','t2','t3','t4','t5'],autopct='%1.1f%%',shadow=True,colors=['yellow','red','pink','cyan','purple'])
plt.title('Pie diagram')
plt.show()
```
